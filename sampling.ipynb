```python
import pandas as pd
import numpy as np
from pycaret.classification import *
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/AnjulaMehto/Sampling_Assignment/main/Creditcard_data.csv')
```

```python
df.shape
```

```python
sum(df['Class'])
from sklearn.preprocessing import normalize
Amount = normalize([df['Amount']])[0]
df['Amount'] = Amount
```

```python
X = df.drop(['Class','Time'], axis=1)
y = df['Class']
```

```python
X
```

# Oversampling using smote

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
print(y.value_counts())
```

```python
X
```

```python
new_df = pd.concat([X, y], axis=1)
new_df
```

## Simple Random Sampling

```python
# simple random sampling
n = int((2*2 * 0.5*0.5)/(0.05**2))
sampled_df = new_df.sample(n=n, random_state=42)
sampled_df
```

```python
setup(data = sampled_df, target='Class')
cm = compare_models()
```

## Systematic Sampling

```python
# systematic sampling
interval = 2
systematic_df = new_df.iloc[::interval]
systematic_df
```

```python
# stratified sampling

n = int((2*2*0.5*0.5)/((0.05)**2))

strata = new_df.groupby('Class')

# sample 2 rows from each stratum
stratified_df = strata.apply(lambda x: x.sample(n))

stratified_df
```

```python
setup(data = systematic_df, target='Class')
cm = compare_models()
```

## Cluster Sampling

```python
# cluster sampling
import numpy as np

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=15, random_state=42).fit(new_df)
cluster_assignments = kmeans.labels_

# Select the clusters you want to include in the sample
selected_clusters = [0, 2, 4, 5, 8]

cluster_series = pd.Series(cluster_assignments)

# Create the new DataFrame containing only the rows from the selected clusters
df_cluster_sample = new_df[cluster_series.isin(selected_clusters)]


# Print the resulting DataFrame
(df_cluster_sample)
```

```python
setup(data = df_cluster_sample, target='Class')
cm = compare_models()
```

```python
convenience_sample = pd.concat([new_df.head(330), new_df.tail(330)])
convenience_sample
```
